```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#dataset
df = pd.read_csv('C:/Users/tirth/OneDrive/Desktop/ecomm-analysis/data/data.csv', encoding='ISO-8859-1')
df.head()
```

```python
df.drop_duplicates(inplace=True)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.dropna(subset=['CustomerID'], inplace=True)
```

```python
df.groupby('Country')['InvoiceNo'].count().sort_values(ascending=False).head(5)
```

```python
# Added a new column for TotalPrice
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df['Month'] = df['InvoiceDate'].dt.to_period('M')

monthly_revenue = df.groupby('Month')['TotalPrice'].sum()

monthly_revenue.plot(kind='line')
```

```python
#top selling products
df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
reference_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalPrice': 'sum'
}).rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalPrice': 'Monetary'
})
rfm.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=1)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
sns.scatterplot(x='Recency', y='Monetary', hue='Cluster', data=rfm)
```

```python
rfm['Cluster'] = rfm['Cluster'].astype(str)  # Optional for better plotting
rfm.groupby('Cluster').mean().round(2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Recency vs Monetary
sns.scatterplot(data=rfm, x='Recency', y='Monetary', hue='Cluster')
plt.title('Recency vs Monetary by Cluster')
plt.show()

# Frequency vs Monetary
sns.scatterplot(data=rfm, x='Frequency', y='Monetary', hue='Cluster')
plt.title('Frequency vs Monetary by Cluster')
plt.show()
```

```python
cluster_map = {
    '0': 'High-Value',
    '1': 'New Customers',
    '2': 'At Risk',
    '3': 'Churned'
}
rfm['Segment'] = rfm['Cluster'].map(cluster_map)
```
